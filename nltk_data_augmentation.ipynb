```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/AI_text Detection/custom_train_data.csv",header=None)
```

```python
from textattack.augmentation import EmbeddingAugmenter, WordNetAugmenter
import textattack

# Initialize augmenters
embedding_augmenter = EmbeddingAugmenter()
wordnet_augmenter = WordNetAugmenter()

def general_text_augmentation(texts):
    augmented_texts = []
    for text in texts:
        # Use multiple augmenters
        augmented_text = embedding_augmenter.augment(text)
        if not augmented_text:
            augmented_text = wordnet_augmenter.augment(text)
        if not augmented_text:
            augmented_text = text
        augmented_texts.extend(augmented_text)

    # Limit the number of augmentations to the desired maximum
    return augmented_texts

# Example usage
texts = [
    "The quick brown fox jumps over the lazy dog.",
    "Natural language processing is a fascinating field.",
    "Machine learning algorithms can be very complex."
]

augmented_texts = general_text_augmentation(texts)

# Print the original and augmented texts
print("Original Texts:")
for text in texts:
    print(text)

print("\nAugmented Texts:")
for text in augmented_texts:
    print(text)
```
